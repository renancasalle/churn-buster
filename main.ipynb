```python
import pandas as pd

file_path = 'cancelamentos.csv'

df = pd.read_csv(file_path)

df = df.drop(columns='CustomerID')

display(df)
```

```python
df = df.dropna()

display(df.info())
```

```python
# qtd de clientes que cancelaram
display(df['cancelou'].value_counts())

#valor em percentual
display(df['cancelou'].value_counts(normalize=True) ) #.map("{:.2%}".format)
```

```python
import plotly.express as px

for col in df.columns:
    fig = px.histogram(df, x=col, color="cancelou")
    fig.show()

fig = px.histogram(df, x="duracao_contrato", color="cancelou")
fig.show()
```

```python
#como ficariam os cancelamentos se os contratos mensais fossem removidos
df = df[df['duracao_contrato']!='Monthly']

display(df['cancelou'].value_counts(normalize=True) ) 
```
